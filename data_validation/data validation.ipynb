```python
import pandas as pd
compounds = pd.read_csv('homePage_data.csv')
pd.set_option('display.max_columns', None)
compounds.head()
```

```python
#validate each compound not located in more than one area
area = {}
for compound_id in compounds['id']:
    if compound_id not in area.keys():
        try:
            area[compound_id] = int(compounds[compounds['id'] == compound_id]['area_id'])
        except Exception as e:
            print(e)
            print("it's not okay only series of single element could parse to int")
#this code in okay then relation should be one two many, many compounds could locate in one area
```

```python
#validate each compound has only one developer
compound_developer = {}
for compound_id in compounds['id']:
    if compound_id not in compound_developer.keys():
        try:
            compound_developer[compound_id] = int(compounds[compounds['id'] == compound_id]['developer_id'])
        except Exception as e:
            print(e)
            print("it's not okay only series of single element could parse to int")
#this code in okay then relation should be one two many, one develpor could develop many compounds
```
